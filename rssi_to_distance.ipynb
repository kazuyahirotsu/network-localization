```python
# -17.2 * log10(distance) - 62.4
# graph of the path loss model
import matplotlib.pyplot as plt
import numpy as np

d = np.arange(1, 10000, 1)
L = -17.2 * np.log10(d) - 62.4

plt.plot(d, L)
plt.xlabel('Distance (m)')
plt.ylabel('Path Loss (dB)')
plt.title('Path Loss Model')
plt.grid(True)
plt.show()
```

```python
# -17.2 * log10(distance) - 62.4 + np.random.normal(0, 3)
# graph of the path loss model with shadowing

d = np.arange(1, 10000, 1)
L = -17.2 * np.log10(d) - 62.4 + np.random.normal(0, 3, len(d))

plt.plot(d, L)
plt.xlabel('Distance (m)')
plt.ylabel('RSSI (dB)')
plt.grid(True)
plt.show()
```

```python
# -17.2 * log10(distance) - 62.4 + np.random.normal(0, 3)
# graph of the path loss model with shadowing

d1 = np.arange(1, 10000, 1)
L1 = -17.2 * np.log10(d) - 62.4 + np.random.normal(0, 3, len(d))
d2 = np.arange(1, 10000, 1)
L2 = -20 * np.log10(d) - 62.4 + np.random.normal(0, 3, len(d))

plt.figure(facecolor='white')
plt.plot(d1, L1, label='Train Model')
plt.plot(d2, L2, label='Test Model')
plt.xlabel('Distance (m)')
plt.ylabel('RSSI (dB)')
plt.legend()
plt.grid(True)
plt.show()
```
